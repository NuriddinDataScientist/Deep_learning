```python
# Kerakli kutubxonalarni yuklab olamiz
from fastai.vision.all import *
from ipywidgets import widgets
```

```python
#data
path = untar_data(URLs.PETS)/'images'

# labellovchi funksiya
def is_cat(x): return x[0].isupper()

# datalarni yuklochi 
dls =  ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid =0.2, seed=42,
    label_func=is_cat, item_tfms = Resize(224)
)

# modelni o'qitish
learn = vision_learner(dls, resnet34, metrics =accuracy)
learn.fine_tune(2)
```

```python
# rasm yuklash uchun widget
upload = widgets.FileUpload()
upload
```

```python
uploaded_file_content = upload.value[0]['content'] 

img = PILImage.create(io.BytesIO(uploaded_file_content))
pred, _, prob = learn.predict(img)
print(f"Mushukmi?: {pred}")
print(f"aniqlilik ehtimolligi: {prob[1].item():.3f}")
img
```
